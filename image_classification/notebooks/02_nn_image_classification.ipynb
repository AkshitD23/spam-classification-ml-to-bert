IMPORTING LIBRARIES

```python
import idx2numpy
import matplotlib.pyplot as plt


from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, precision_score, recall_score
```

GETTING DATA

```python
x_train = idx2numpy.convert_from_file("../MNIST/train-images.idx3-ubyte")
y_train = idx2numpy.convert_from_file("../MNIST/train-labels.idx1-ubyte")

x_test = idx2numpy.convert_from_file("../MNIST/t10k-images.idx3-ubyte")
y_test = idx2numpy.convert_from_file("../MNIST/t10k-labels.idx1-ubyte")

```

IMAGE VISUALS

```python
plt.imshow(x_test[90])
plt.title(y_test[90])
plt.show()
```

FLATTENING THE DATA

```python
x_train = x_train.reshape(-1,784)
x_test = x_test.reshape(-1,784)
```

SCALING DATA

```python
x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
nn_model = MLPClassifier(
    hidden_layer_sizes=(128,),
    max_iter = 500,
    random_state = 42 
)
```

```python
nn_model.fit(x_train,y_train)
```

```python
nn_pred = nn_model.predict(x_test)
```

```python
confusionMatrix_nn = confusion_matrix(y_test,nn_pred)
display = ConfusionMatrixDisplay(confusionMatrix_nn)
display.plot()
plt.show()
```

```python
precision = precision_score(y_test,nn_pred,average = None)
print(precision)
```

```python
recall = recall_score(y_test,nn_pred,average = None)
print(recall)
```
